## Accuracy calculating function

````python
from rouge_score import rouge_scorer
````

````python


def calculate_accuracy(reference_summary, generated_summary, model_name):
    generated_summary = generated_summary.replace('\n', ' ')
    reference_summary = reference_summary.replace('\n', ' ')

    # Initialize the ROUGE scorer
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

    # Calculate ROUGE scores
    scores = scorer.score(reference_summary, generated_summary)

    print(f'Model name: {model_name}')
    print("ROUGE-1:", scores['rouge1'])
    print("ROUGE-2:", scores['rouge2'])
    print("ROUGE-L:", scores['rougeL'])

    # Prepare the scores to return
    return {
        'model_name': model_name,
        'rouge1': scores['rouge1'],
        'rouge2': scores['rouge2'],
        'rougeL': scores['rougeL']
    }


# # Calculate and store the scores
# scores = calculate_accuracy(reference_summary, generated_summary, model_name)

# # Access scores for comparison
# print(scores)
````

````python
import time
from transformers import BigBirdPegasusForConditionalGeneration, AutoTokenizer
import torch

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("google/bigbird-pegasus-large-arxiv")
model = BigBirdPegasusForConditionalGeneration.from_pretrained("google/bigbird-pegasus-large-arxiv")

# Function to split text into smaller chunks based on token length
def split_text_into_chunks(text, max_length):
    tokens = tokenizer.encode(text, truncation=False)
    print(f"Total number of tokens: {len(tokens)}")  # Debugging: print the total number of tokens
    chunks = [tokens[i:i + max_length] for i in range(0, len(tokens), max_length)]
    for i, chunk in enumerate(chunks):
        print(f"Chunk {i+1} length: {len(chunk)}")  # Debugging: print the length of each chunk
    return chunks

# Split the text into chunks
max_chunk_length = 4096  # Ensure this is less than or equal to the model's max length
chunks = split_text_into_chunks(text_i_str, max_chunk_length)

# Start the timer
start_time = time.time()

# Generate summaries for each chunk
summaries = []
for i, chunk in enumerate(chunks):
    chunk = chunk[:max_chunk_length]  # Ensure each chunk is within the model's maximum length
    input_ids = torch.tensor([chunk])
    print(f"Input IDs length for chunk {i+1}: {len(input_ids[0])}")  # Debugging: print the length of input IDs
    input_ids = input_ids.to(torch.int64)  # Ensure the tensor is of the correct dtype
    attention_mask = torch.ones(input_ids.shape, dtype=torch.long)  # Create an attention mask

    output = model.generate(input_ids, attention_mask=attention_mask, max_length=128, num_beams=5, early_stopping=True)
    summary = tokenizer.decode(output[0], skip_special_tokens=True)
    summaries.append(summary)
    print(f"Processed chunk {i+1}/{len(chunks)}")

# Stop the timer
end_time = time.time()

# Combine the summaries
final_summary_iv = " ".join(summaries)

# Calculate and print the total time consumed
total_time = end_time - start_time
print(f"Total time consumed: {total_time:.2f} seconds")
````

## Comparing model performences

````python
import pandas as pd
from tabulate import tabulate


# Collect data from variables
models_data = [scores_i, scores_ii, scores_iii, scores_iv]

# Extract relevant data for the DataFrame
data = []
for model in models_data:
    model_name = model['model_name']
    rouge1_f1 = model['rouge1'][2]  # F1 score is the third element
    rouge2_f1 = model['rouge2'][2]  # F1 score is the third element
    rougeL_f1 = model['rougeL'][2]  # F1 score is the third element
    data.append({'model_name': model_name, 'rouge1_f1': rouge1_f1, 'rouge2_f1': rouge2_f1, 'rougeL_f1': rougeL_f1})

# Create a DataFrame
df = pd.DataFrame(data)

# Determine the best scores
best_rouge1_model = df.loc[df['rouge1_f1'].idxmax()]
best_rouge2_model = df.loc[df['rouge2_f1'].idxmax()]
best_rougeL_model = df.loc[df['rougeL_f1'].idxmax()]

# Print the DataFrame using tabulate for better formatting
print(tabulate(df, headers='keys', tablefmt='psql'))

# Print the best models
print("\nBest performing models:")
print(f"ROUGE-1 best model: {best_rouge1_model['model_name']} with F1 score {best_rouge1_model['rouge1_f1']}")
print(f"ROUGE-2 best model: {best_rouge2_model['model_name']} with F1 score {best_rouge2_model['rouge2_f1']}")
print(f"ROUGE-L best model: {best_rougeL_model['model_name']} with F1 score {best_rougeL_model['rougeL_f1']}")
````

## What next??

To improve the summarization of scientific papers from arXiv, you can consider several approaches. Here are a few strategies to enhance the quality of your summaries:

- Fine-tuning the Model:

Fine-tune a pre-trained summarization model on a dataset of scientific papers to make it more specialized in understanding and summarizing scientific content.
- Experimenting with Different Hyperparameters:

Adjust parameters like max_length, num_beams, and length_penalty to optimize the quality of the summaries.
- Using Ensembles:

Combine summaries from multiple models to produce a more comprehensive and accurate summary.
- Incorporating Additional Context:

Use more context from the paper, such as the abstract, introduction, and conclusion, to provide a richer summary.
- Post-Processing:

Refine the generated summaries using post-processing techniques like removing redundant sentences or ensuring the coherence of the summary.
